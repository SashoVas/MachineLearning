```python
import numpy as np
def random_walk():
    step=0
    steps=[step]
    for i in range(100):
        dice=np.random.randint(2,7)
        if dice<3:
            if step>0:
                step-=1
        elif dice<6:
            step+=1
        else:
            step+=np.random.randint(2,7)
        steps.append(step)
    return steps
```

```python
import matplotlib.pyplot as plt

plt.suptitle('Random walk')
plt.plot(list(range(101)),steps)

plt.show()
```

```python
def plot_walks(num_walks,walk_func):
    plt.suptitle('Random walks')
    np.random.seed(123)
    for i in range(num_walks):
        plt.plot(list(range(101)),walk_func())

    plt.show()
```

```python
import numpy as np
def random_walk_with_fall():
    step=0
    steps=[step]

    for i in range(100):
        dice=np.random.randint(2,7)
        fall=np.random.rand()
        if dice<3:
            if step>0:
                step-=1
        elif dice<6:
            step+=1
        else:
            step+=np.random.randint(2,7)

        if fall<0.005:
            step=0
        steps.append(step)
    return steps
```

```python
plot_walks(20,random_walk_with_fall)
```

```python

plt.suptitle('Random walks')
plt.hist([random_walk_with_fall()[-1] for i in range(500)])

plt.show()
```
